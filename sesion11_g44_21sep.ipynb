```python
#librerias basicas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#librerias especializadas
from statsmodels.tsa.seasonal import seasonal_decompose

#tama√±o grafica (largo,alto)
plt.rcParams['figure.figsize'] = (10,6)
```

```python
## datos
pasajeros = pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/airline_passengers.csv',index_col = 'Month', parse_dates= True)
```

```python
pasajeros.head()
```

```python
#renombrar
pasajeros = pasajeros.rename(columns={'Thousands of Passengers':'miles de pasajeros'})
```

```python
pasajeros.head()
```

```python
#siempre se grafican las series de tiempo
pasajeros['miles de pasajeros'].plot()
plt.title('Pasajeros mensuales E.U 1949-1960')
plt.ylabel('miles')
plt.xlim('1954-01','1954-12')
plt.xlabel('mes');
```

La serie tiene tendencia creciente, componente estacional.

Los pasajeros siempre se incrementan en los meses junio-agosto



##Modelo ETS(Error-trend-seasonally)
Modelo Irregular-Tendencia(ciclo)-Estacional



```python
from operator import add
ets = seasonal_decompose(pasajeros['miles de pasajeros'], model='mul')
```

```python
ets.plot()
plt.show()
#la de arriba es la serie original
```

```python
ets.trend.plot();
```

```python
ets.seasonal.plot();
```

```python
#desestacionalizar la serie (quitarle el componente estacional)
pasajeros['desestacionalizada'] = pasajeros['miles de pasajeros']/ets.seasonal
```

```python
pasajeros['desestacionalizada'].plot();
```
