```python
#ticker(nombre en bolsa) walmart de mexico
walmex = yf.download('WALMEX.MX',
                     start = '2015-01-01',
                     end = '2019-12-31',
                     progress =False)
```

```python
walmex.head()
```

```python
precio_cierre = walmex[['Adj Close']]
```

```python
precio_cierre.head()
```

```python
precio_cierre.to_excel('precio_cierre.xlsx', index = True)
```

```python
#calculo de promedio movil simple
precio_cierre['PMS-7']=precio_cierre['Adj Close'].rolling(7).mean()
precio_cierre['PMS-50']=precio_cierre['Adj Close'].rolling(50).mean()
```

```python
precio_cierre.head()
```

```python
precio_cierre['retorno'] =precio_cierre['Adj Close']-precio_cierre['Adj Close'].shift(1)
```

```python
precio_cierre.head()
```

```python
precio_cierre['estrategia'] =np.select(precio_cierre['PMS-7']>=precio_cierre['PMS-50'],'mantener','vender')
```
