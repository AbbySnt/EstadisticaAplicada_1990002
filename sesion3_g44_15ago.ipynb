## Funciones de Pandas

```python
#Sort_values, por default es de menor a mayor
df_clase.sort_values('Edad')
```

```python
#Ordenar de mayor a menor
df_clase.sort_values('Edad',ascending=False)
```

```python
#ordenar por dos columnas
df_clase.sort_values(['Nombre','Edad'])
```

```python

df_clase['Edad'].mean()
```

```python
df_clase['Edad'].max()
```

```python
#np.where (condicion,valor si se cumple, valor si no se cumple)
import numpy as np
```

```python
df_clase['mayoria_edad']=np.where(df_clase['Edad']>=18,'Mayor','Menor')
```

```python
df_clase
```

```python
#np.select (condiciones,valor para cada condicion,valor si no se cumple ninguna condicion)

condiciones = [df_clase['Edad']<=17,
               df_clase['Edad']<23,
               df_clase['Edad']>=23]
valores =['Prepa','Universidad','Trabajo']

df_clase['Situacion'] = np.select(condiciones,valores,default = 'Invalido')
```

```python
df_clase
```

```python
#contar
df_clase.count()
```

```python
df_clase['mayoria_edad'].count()
```

```python
df_clase.groupby('mayoria_edad').count()
```

```python
df_clase.groupby('mayoria_edad').mean(numeric_only=True).round(1)
```
