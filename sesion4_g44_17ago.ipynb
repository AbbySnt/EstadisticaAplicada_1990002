```python
#eliminar una columna
calificaciones = calificaciones[['Nombre','Apellido','Calficaci칩n Final']]
```

```python
#renombrar el nombre de una columna
calificaciones = calificaciones.rename(columns={'Calficaci칩n Final':'CalFinal'})
```

```python
calificaciones.head()
```

```python
#calcular el promedio
calificaciones['CalFinal'].mean()
```

```python
#clasificar que alumnos aprobaron
calificaciones['situacion'] = np.where(calificaciones['CalFinal']>=70,'Aprobado','Reprobado')
```

```python
calificaciones.head()
```

```python
#contar cuandot alumnos estan aprobados y reprobados
calificaciones.groupby('situacion')['CalFinal'].count()
```

```python
#promedio de aprobados y reprobados
calificaciones.groupby('situacion')['CalFinal'].mean().round(1)
```

```python
#contarlos y sacar el promedio al mismo tiempo
## agg sirve para decirle a python que se la va a indicar varias cosas
calificaciones.groupby('situacion')['CalFinal'].agg(['mean','count','std']).round(1)
```

```python
aprobados = calificaciones.query('situacion == "Aprobado" ')
reprobados = calificaciones.query('situacion == "Reprobado" ')
```

## Matplotlib mathematical plot library

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
 #matplotlib es para que siempre se haga un grafica se muestre
```

```python
personas = [1,2,3,4]
edades = [25,27,21,30]
```

```python
#graficar
plt.bar(personas,edades)
plt.show() #forma correcta  forma incorrecta: es poner un ; al final del plt
```

```python
#darle los ejes y titulo y color
plt.bar(personas,edades, color = 'yellow')
plt.xlabel('personas')
plt.ylabel('edades (a침os)')
plt.title('Edades de 4 desconocidos')
plt.show()
```

```python
#darle los ejes y titulo y color para cada barra
plt.bar(personas,edades, color = ['yellow','green','red','orange'])
plt.xlabel('personas')
plt.ylabel('edades (a침os)')
plt.title('Edades de 4 desconocidos')
plt.show()
```
