```python
## librerias de cajon
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## datos con los que se van a trabajar
edades = [23,25,30,19]
personas = ['Juan','Pedro','Ana','Lissa']

dicc_1={'municipio':['Escobedo','Apodaca','Monterrey','Santiago'],'nacimientos':[200,350,400,120],'muertes':[20,35,40,18],'matrimonios':[56,68,78,19]}
datos_mun = pd.DataFrame(dicc_1)
```

```python
datos_mun
```

```python
edades, personas
```

```python
## Graficas
plt.bar(personas,edades)
plt.title('edades de 4 personas')
plt.xlabel('personas')
plt.ylabel('edades')
plt.show
```

```python
#creacion de grafico en base a dataframes
plt.bar(datos_mun['municipio'],datos_mun['matrimonios'])
plt.title('Matrimonios en N.L')
plt.xlabel('municipio')
plt.ylabel('cantidad de matrimonio')
```

```python
#dos graficas juntas
#fig es para darle nombre al pizarron fig,(nombre de la grafica. nombre de la segunda grafica)=plt.subplots(filas#,columnas#)
fig,(graf1,graf2) =plt.subplots(2,1)
#grafica 1
graf1.bar(personas,edades)
#grafica 2
graf2.bar(datos_mun['municipio'],datos_mun['matrimonios'])
plt.show()
```
